```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import textract
```

```python
with open('stop_words.txt') as file:
    stopwords = {line.strip() for line in file}
```

```python
planogoverno_txt = textract.process("pdfs/plano_governo_JoaoGoulart.pdf").decode('utf-8')
with open('txt/plano_governo_JoaoGoulart.txt', 'w') as file:
    file.write(planogoverno_txt)
```

```python
wordcloud = WordCloud(min_font_size = 20, 
                      max_font_size = 300,
                      width = 2000,
                      height = 1000,
                      mode = "RGB",
                      normalize_plurals = True,
                      stopwords = stopwords).generate(planogoverno_txt)
plt.figure(figsize = (16,9))
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.savefig('pics/wordcloud_JoaoGoulart.png', bbox_inches='tight', transparent = True, pad_inches = 0)
plt.show()
```
