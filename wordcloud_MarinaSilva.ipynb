```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import codecs
import textract
```

```python
toText = textract.process("pdfs/plano_governo_Marina.pdf")
f = open('txt/plano_governo_Marina.txt', 'w')
f.write(toText)
f.close
```

```python
stopwords = set(STOPWORDS)
new_words = []
with open('stop_words.txt', 'r') as sf:
    [new_words.append(word) for line in sf for word in line.split()]
new_stopwords = stopwords.union(new_words)
```

```python
planogoverno_txt = codecs.open('txt/plano_governo_Marina.txt', encoding='utf-8').read()
wordcloud = WordCloud(min_font_size=20, 
                      max_font_size=300,
                      width = 2000,
                      height = 1000,
                      mode="RGB",
                      normalize_plurals=True,
                      stopwords=new_stopwords).generate(planogoverno_txt)
plt.figure(figsize=(16,9))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
