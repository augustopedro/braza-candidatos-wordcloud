```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import textract
import codecs
```

```python
with open('stop_words.txt') as file:
    stopwords = {line.strip() for line in file}
```

```python
with open("txt/geral_plano_governo.txt") as myFile:
    planogoverno_txt = myFile.read()
```

```python
wordcloud = WordCloud(min_font_size = 20, 
                      max_font_size = 300,
                      width = 2000,
                      height = 1000,
                      mode = "RGB",
                      normalize_plurals = True,
                      stopwords = stopwords).generate(planogoverno_txt)
plt.figure(figsize = (16,9))
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.savefig('pics/wordcloud_geral.png', bbox_inches='tight', transparent = True, pad_inches = 0)
plt.show()
```
