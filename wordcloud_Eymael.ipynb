```python
from wordcloud import WordCloud, STOPWORDS
from os import path
import matplotlib.pyplot as plt
import textract
import codecs
```

```python
stopwords = set(STOPWORDS)
new_words = []
with codecs.open('stop_words.txt', 'r', encoding = 'utf-8') as sf:
    [new_words.append(word) for line in sf for word in line.split()]
new_stopwords = stopwords.union(new_words)
```

```python
toText = textract.process("pdfs/plano_governo_Eymael.pdf")
f = open('txt/plano_governo_Eymael.txt', 'w')
f.write(toText)
f.close
```

```python
planogoverno_txt = codecs.open('txt/plano_governo_Eymael.txt', encoding = 'utf-8').read()
wordcloud = WordCloud(min_font_size = 20, 
                      max_font_size = 300,
                      width = 2000,
                      height = 1000,
                      mode = "RGB",
                      normalize_plurals = True,
                      stopwords = new_stopwords).generate(planogoverno_txt)
plt.figure(figsize = (16,9))
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.savefig('pics/wordcloud_Eymael.png', bbox_inches='tight', transparent = True, pad_inches = 0)
plt.show()
```
